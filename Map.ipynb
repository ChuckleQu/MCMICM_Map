```python
data = pd.read_excel('2021MCMProblemC_DataSet.xlsx')
positive_data = data.loc[data['Lab Status'] == "Positive ID"]
positive_latitude = positive_data['Latitude'].values
print(positive_latitude)
positive_longitude = positive_data['Longitude'].values
print(positive_latitude)
```

```python
city_map = folium.Map(location=[48.9830, -123.2112], zoom_start=8)
for i in range(len(positive_latitude)):
    folium.Circle(
    radius=30*1000,
    location=(positive_latitude[i], positive_longitude[i]),
    color='#3388ff',
    fill=True,
    ).add_to(city_map)

city_map.add_child(folium.LatLngPopup())
city_map
```
